```python
# importing lib and loading dataset
import pandas as pd
dataset= pd.read_csv('CKD.csv')
dataset
```

```python
# check the dataset for null vlaues
dataset.isna().sum()
```

```python
# to get the col names
dataset.columns
```

```python
# drop the unwanted col names
df =dataset.drop(['su','pc','pcc','bgr','sod','pot','htn','dm','cad','ba','bu','pcv'],axis = 1)
```

```python
df
```

```python
# changing categorical value to numeric value

df = pd.get_dummies(df, drop_first=True)
df = df.astype(int)
df
```

```python
# to find the unique value
df['classification_yes'].value_counts()
```

```python
df.columns
```

```python
# rename the column
df =df.rename(columns ={'classification_yes':'Classification'})
```

```python
df.columns
```

```python
# to split the input and output
independent =df[['age', 'bp', 'al', 'sc', 'hrmo', 'wc', 'rc', 'sg_b', 'sg_c', 'sg_d',
                 'sg_e', 'rbc_normal', 'appet_yes', 'pe_yes', 'ane_yes']]
dependent = df[['Classification']]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(independent,dependent,test_size =0.30, random_state = 0)
X_train
```

```python
X_test
```

```python
Y_test
```

```python
from sklearn.ensemble import RandomForestClassifier
classifier = RandomForestClassifier(n_estimators=100,criterion='gini',max_features='sqrt')
classifier = classifier.fit(independent, dependent)
classifier
```

```python
test_pred = classifier.predict(X_test)
test_pred
```

```python
from sklearn.metrics import r2_score
r2 = r2_score(Y_test,test_pred)
r2
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
parameters = {'n_estimators':[10,50,100],
             'criterion':['gini','entropy','log_loss'],
             'max_features':['sqrt','log2']}
grid = GridSearchCV(RandomForestClassifier(), parameters, refit =True, verbose = 3, n_jobs = -1)
grid.fit(independent,dependent)
```

```python
print(grid.best_params_)
```

```python
classifier = grid.cv_results_
classifier
print("The R score value for best paramater: ",format(grid.best_params_))
```

```python
table = pd.DataFrame.from_dict(classifier)
table
```
