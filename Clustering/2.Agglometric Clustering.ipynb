```python
#importing library
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#load the dataset
dataset = pd.read_csv("Mall_Customers.csv")
dataset
```

```python
# X = dataset.iloc[0:5,0:4].values #[r:c] rows and cols [start:end,start:end]
# X= dataset.iloc[:,3:5]
X= dataset.iloc[:,3:5].values #retrive the data based on location
print(X)
```

```python
import scipy.cluster.hierarchy as dg
dendrogram = dg.dendrogram(dg.linkage(X,method='ward'))
plt.title("Dendrogram")
plt.xlabel("Customers")
plt.ylabel("Euclidean distance")
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
model = AgglomerativeClustering(n_clusters =5)
label = model.fit_predict(X)
label
```
