```python
#importing library
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#load the dataset
dataset = pd.read_csv("Mall_Customers.csv")
dataset
```

```python
# X = dataset.iloc[0:5,0:4].values #[r:c] rows and cols [start:end,start:end]
# X= dataset.iloc[:,3:5]
X= dataset.iloc[:,3:5].values #retrive the data based on location
print(X)
```

```python
# to find the cluster 
from sklearn.cluster import KMeans
list1 =[] # empty array
for i in range(1,11): 
    kmeans = KMeans(n_clusters =i,init='k-means++',random_state=30) # take the parameter from sklearn page
    kmeans.fit(X)
    list1.append(kmeans.inertia_) # need to find inertia to find the no of cluster
plt.plot(range(1,11),list1)
plt.title("The Elbow Method")
plt.xlabel("No of cluster")
plt.ylabel("WCSS")
plt.show()
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters =5,init='k-means++',random_state=30) # now put the cluster value as found above
y_kmeans = kmeans.fit_predict(X) # fit_predict is to find the clsuter for the data.
```
