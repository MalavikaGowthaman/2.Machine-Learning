```python
#importing library
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#load the dataset
dataset = pd.read_csv("Mall_Customers.csv")
dataset
```

```python
# X = dataset.iloc[0:5,0:4].values #[r:c] rows and cols [start:end,start:end]
# X= dataset.iloc[:,3:5]
X= dataset.iloc[:,3:5].values #retrive the data based on location
print(X)
```

```python
from sklearn.cluster import MeanShift
Ms= MeanShift(bandwidth= 17).fit(X)
y_Ms=Ms.fit_predict(X)
```

```python
y_Ms
```

```python
supervised = dataset # to transform the data into supervised
```

```python
supervised['Cluster_group']=y_Ms # to add the new col
supervised
```

```python
import seaborn as sns
Graph = sns.lmplot(data=supervised, x= supervised.columns[3], y= supervised.columns[4],
                  hue= supervised.columns[5],fit_reg=False) # in plot need to put data, X and Y axis. hue is for colours.
```
