```python
# importing library
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#loading dataset
dataset = pd.read_csv("insurance_pre.csv")
dataset
```

```python
#transform categorical value to numerical value
dataset = pd.get_dummies(dataset, drop_first =True)
dataset = dataset.astype(int)
dataset
```

```python
dataset.columns
```

```python
independent = dataset[['age', 'bmi', 'children','sex_male', 'smoker_yes']]
dependent = dataset[['charges']]
```

```python
#split into training set and test
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(independent, dependent, test_size = 1/3, random_state = 0)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
from sklearn.svm import SVR #importing svr 
from sklearn.model_selection import GridSearchCV # importing gridsearch to make the crossvalidation 
parameters = {'kernel':['linear', 'poly', 'rbf', 'sigmoid'], # paramters are svr from website
              'C':[1,10,100]}
#               'gamma':['scale', 'auto']}

grid = GridSearchCV(SVR(),parameters, refit = True, verbose =3, n_jobs=-1) #refit true is for finding the best model,
                                                                           #false will give last model
grid.fit(independent,dependent) # can able to call directly without training and testing 
```

```python
print(grid.best_params_)
```

```python
regressor =grid.cv_results_ 
regressor
print('The R score value for best parameter: ',format(grid.best_params_))
```

```python
table = pd.DataFrame.from_dict(regressor)
table
```

```python
age = float(input("Age: "))
bmi = float(input("BMI: "))
children = int(input("No of Children: "))
sex_male= int(input("Sex Male :0 , Female: 1: "))
smoker_yes= int(input("Smoker: 0, Non-smoker:1: "))
```

```python
Final_prediction =grid.predict([[age,bmi,children,sex_male,smoker_yes]])
print("Final_prediction of SVR: {}".format(Final_prediction))
```
