```python
#import library
import pandas as pd
```

```python
dataset = pd.read_csv("50_Startups.csv")
dataset
```

```python
# here data have categorical value 
# need to change state col value into numerical using one hot encoding

# dataset=pd.get_dummies(dataset)
# dataset
dataset=pd.get_dummies(dataset,drop_first=True) # to reduce the repetative info need to use drop 

dataset=dataset.astype(int) # to change true or flase to int
dataset
```

```python
# to get column names from table
dataset.columns 
```

```python
# need to assign input and output
independent = dataset[['R&D Spend', 'Administration', 'Marketing Spend','State_Florida', 'State_New York']]
print(independent)
dependent=dataset[['Profit']]
print(dependent)
```

```python
# spliting training and testing data

from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(independent,dependent,test_size=0.30,random_state=0)
X_train,X_test,Y_train,Y_test
```

```python
#Model creation
from sklearn.tree import DecisionTreeRegressor
regressor=DecisionTreeRegressor(criterion='poisson',splitter='best')
regressor=regressor.fit(X_train,Y_train)
```

```python
# ploting graph
import matplotlib.pyplot as plt
from sklearn import tree
tree.plot_tree(regressor)
plt.show()
```

```python
# testing
test_pred= regressor.predict(X_test)
test_pred
```

```python
#Evalution Metrics
from sklearn.metrics import r2_score
r2 = r2_score(Y_test,test_pred)
```

```python
r2
# r2 should be nearer to 1 here it is not performing good for this dataset
```
